#**Importacion de datos**

````python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image, display
from matplotlib.ticker import FuncFormatter
````

````python
url_1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url_2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url_3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url_4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda_1 = pd.read_csv(url_1)
tienda_2 = pd.read_csv(url_2)
tienda_3 = pd.read_csv(url_3)
tienda_4 = pd.read_csv(url_4)
````

##**Comparación de ingresos de tiendas**

````python
ingresos_tiendas = {"Nombre tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4" ],
                       "Ingresos": [ingreso_total_tienda_1, ingreso_total_tienda_2, ingreso_total_tienda_3, ingreso_total_tienda_4]}
df_ingresos_tiendas = pd.DataFrame(ingresos_tiendas)
df_ingresos_tiendas = df_ingresos_tiendas.sort_values(by="Ingresos", ascending=True)
df_ingresos_tiendas
````

###**Tienda con mayor ingresos**

````python
maximo = max(df_ingresos_tiendas['Ingresos'])
valor_maximo = df_ingresos_tiendas.loc[df_ingresos_tiendas['Ingresos'] == maximo]
valor_maximo
````

###**Tienda con menores ingresos**

````python
minimo = min(df_ingresos_tiendas['Ingresos'])
valor_minimo = df_ingresos_tiendas.loc[df_ingresos_tiendas['Ingresos'] == minimo]
valor_minimo
````
